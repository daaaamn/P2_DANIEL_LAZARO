```python
import pandas as pd
import numpy as np
import mpld3
import matplotlib.pyplot as plt
import plotly.tools as tls
import plotly.io as pio
from scipy.spatial import ConvexHull

coordenadas = df[['Longitud', 'Latitud']].values

envolvente = ConvexHull(coordenadas)

fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(coordenadas[:, 0], coordenadas[:, 1], c='blue', label="Puntos originales")  

for simplex in envolvente.simplices:
    ax.plot(coordenadas[simplex, 0], coordenadas[simplex, 1], 'r-', linewidth=2) 

ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")
ax.set_title("Envolvente Convexa (Convex Hull)")
ax.legend()
ax.grid()

mpld3.save_html(fig, "grafica_envolvente_convexa.html")
```
